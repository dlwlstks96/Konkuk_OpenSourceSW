```python
userInput = input('입력하시오 : ')
#입력된 문자열 분리
arg = userInput.split(' ')

if(arg[0]=='cp'):
    #문자열에 파일 확장명 달아주기
    openFile = arg[1]+'.txt'
    writeFile = arg[2]+'.txt'

    #file1 읽기
    f1 = open(openFile, 'r')
    lines = f1.readlines()
    f1.close()

    #file2 생성 및 적기
    f2 = open(writeFile, 'w')

    for line in lines:
        f2.write(line)
    f2.close()
```

```python
userInput = input('입력하시오 : ')
#입력된 문자열 분리
arg = userInput.split(' ')

if(arg[0]=='wc'):
    lineCount = 0
    wordCount = 0
    #문자열에 파일 확장명 달아주기
    openFile = arg[1]+'.txt'
    
    #파일 읽어 line, word 수 카운트
    f1 = open(openFile, 'r')
    while 1:
        line = f1.readline()
        if not line: break
        lineCount = lineCount + 1
        wordCount += len(line.split(' '))
            
    f1.close()
    
print('라인의 갯수 : {0}, 단어의 갯수 : {1}'.format(lineCount, wordCount))
```

```python
# 1 + 2 - 3 * 4 + 5
#스택 클래스 구현
class stack:
    
    #item을 담을 배열 생성
    def __init__(self):
        self.arr = []
    
    def push(self, item):
        return self.arr.append(item)
    
    def pop(self):
        return self.arr.pop()
    
    def isEmpty(self):
        return self.arr is None
    
    def size(self):
        return len(self.arr)

userInput = input('계산식 입력 : ')

#입력식을 공백으로 분리
spl = userInput.split(' ')

#숫자와 연산자를 담을 스택 생성
Numstack = stack()
OPstack = stack()
#카운트를 하여 숫자와 연산자를 구분하여 스택에 push
count = 1
#push할 연산자가 * 혹은 / 인지 체크하기 위한 bool
opBool = False
#push할 연산자가 - 인지 체크하기 위한 bool
minusCheck = False

'''
1. 사칙연산 규칙을 행하기 위해 Numstack에 숫자를 넣을시
가장 최근에 넣은 OPstack의 연산자를 항상 체크한다
1-1. 만약 가장 최근에 * 혹은 / 연산자를 넣었다면 해당 연산자의
앞 뒤 피연산자를 이용해 연산자에 맞게 계산 후 Numstack에 넣어준다

2. 연산자를 넣을 차례에 - 연산자라면 OPstack에 + 연산자를 넣고
다음 피연산자의 부호를 바꿔 넣어준다

3. 이 과정을 통해 Numstack에는 양수 혹은 음수의 피연산자들이,
OPstack에는 + 연산자만 남게 된다
'''

for x in spl:
    #이전에 담은 연산자가 * 혹은 / 일 경우 if문 실행
    if(opBool==True):
        #연산자 앞 뒤의 피연산자를 이용해 계산 후 Numstack에 다시 push
        Numstack.push(int(x))
        num1 = Numstack.pop()
        num2 = Numstack.pop()
        op = OPstack.pop()
        if(op=='*'):
            Numstack.push((num1*num2))
            opBool = False
            count = count + 1
        elif(op=='/'):
            Numstack.push(num1*num2)
            opBool = False
            count = count + 1
    #이전에 담은 연산자가 * 혹은 / 가 아닐 경우
    else:
        #피연산자를 Numstack에 대입할 차례
        if(count%2==1):
            #이전에 넣은 연산자가 - 일 경우
            if(minusCheck==True):
                #피연산자의 부호를 바꿔 push 한다
                x = -int(x)
                Numstack.push(int(x))
                count = count + 1
                #이후 체크에도 활용하기 위해 다시 bool 값 변경
                minusCheck=False
            #이전에 넣은 연산자가 - 가 아닐 경우 그대로 피연산자 push
            else:
                Numstack.push(int(x))
                count = count + 1
        #연산자를 OPstack에 대입할 차례
        elif(count%2==0):
            #연산자가 - 일 경우 OPstack에 + 를 넣고 체크를 위해 bool 값 변경
            if(x=='-'):
                minusCheck = True
                OPstack.push('+')
                count = count + 1
            #연산자가 - 가 아닐 경우 그대로 연산자 push
            else:
                OPstack.push(x)
                count = count + 1
            #연산자가 * 혹은 / 일 경우 다음 피연산자를 넣을 차례에
            #위에 짜놓은 단계를 밟기 위해 bool값 변경
            if(x=='*' or x=='/'):
                opBool = True
                
# + 연산자만 남은 OPstack과 Numstack을 이용해
#순차적으로 pop을 하며 계산
while(True):
    #Numstack에 피연산자가 하나 남을 경우에는 if문 실행 않고 while문 탈출
    if(Numstack.size()!=1):
        num1 = Numstack.pop()
        num2 = Numstack.pop()
        op = OPstack.pop()
        if(op=='+'):
            Numstack.push(num1+num2)
    else:
        break
        
#결과 출력을 위해 공백 없이 입력식을 이어 붙여 출력식 출력
outputStr = ''
for x in spl:
    outputStr += x

#위 과정을 거치면 Numstack에는 계산 결과값 하나가 남기에 pop과 동시에 호출
result = Numstack.pop()
print('계산식 출력 : {0}={1}'.format(outputStr, result))
```
