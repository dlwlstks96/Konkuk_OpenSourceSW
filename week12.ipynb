```python
import csv

max_temp = -999
max_date = ''
min_temp = 999
min_date = ''
tempGad_temp = 0
tempGap_date= ''

#최고 기온이 가장 높은 날 찾기
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
for row in data:
    if row[-1]=='':
        row[-1] = -999
    row[-1] = float(row[-1])
    if(max_temp<row[-1]):
        max_date = row[0]
        max_temp = row[-1]
f.close()

#최저 기온이 가장 낮은 날 찾기
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
for row in data:
    if row[-2]=='':
        row[-2] = 999
    row[-2] = float(row[-2])
    if(min_temp>row[-2]):
        min_date = row[0]
        min_temp = row[-2]
f.close()

#일교차가 가장 큰 날 찾기
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
for row in data:
    if((row[-2]=='')or(row[-1]=='')): #비어 있는 데이터 처리
        continue
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    tmp = row[-1] - row[-2] #해당일 일교차 구하기
    if(tempGad_temp<tmp): #일교차 비교
        tempGap_date = row[0]
        tempGad_temp = tmp
f.close()

print('일교차가 가장 큰 날')
print(tempGap_date)
print('최저 기온이 가장 낮은 날')
print(min_date)
print('최고 기온이 가장 높은 날')
print(max_date)
```

```python
import csv
import operator

nowLine = '' #현재 처리하고 있는 data line의 호선을 알기 위함
prevLine = '1호선' #이전에 처리한 data line의 호선을 담기 위함

passengerCount = {} #지하철역과 이용객수를 담기 위한 딕셔너리


f = open('subwayfee.csv')
data = csv.reader(f)
next(data)
for row in data: #data 한 줄씩 읽기
    nowLine = row[1] #현재 처리하고 있는 data line의 호선 정보
    if(nowLine != prevLine): #현재 처리 중인 호선 정보가 이전 처리한 호선 정보와 다를 경우
                             # 새로운 호선을 읽기 시작했단 의미로 이전 호선의 실습 결과를 출력
        sdict = sorted(passengerCount.items(), key=operator.itemgetter(1), reverse=True) #이용객 수 내림차순 정렬
        print('<'+prevLine+'>')
        print('이용객이 가장 많은 역 순위')
        count = 1 #순위를 출력하기 위함
        for i in sdict[0:3]: #상위 3개 출력
            print(count,'.',i[0],'(',i[1],')')
            count = count + 1
        print('\n이용객이 가장 적은 역 순위')
        sdict = sorted(passengerCount.items(), key=operator.itemgetter(1), reverse=False) #이용객 수 오름차순 정렬
        count = 1
        for i in sdict[0:3:]: #하위 3개 출력
            print(count,'.',i[0],'(',i[1],')')
            count = count + 1
        print('\n')
        passengerCount = {} #새로운 호선의 이용객 정보를 담기 위해 딕셔너리 초기화
        if(',' in row[-1]): #이용객 수를 str에서 int로 바꾸기 위해 ',' 문자 제거
            row[-1] = row[-1].replace(',','')
        if(',' in row[-2]):
            row[-2] = row[-2].replace(',','')
        row[-1] = int(row[-1]) # 승하차 인원 수 계산하여 딕셔너리에 저장
        row[-2] = int(row[-2])
        tmp = row[-1] + row[-2]
        passengerCount[row[3]] = tmp
        prevLine = row[1] # 다음 호선 정보와 비교하기 위해 현재의 호선 정보를 prevLins에 저장
    else: #현재 처리 중인 호선 정보가 이전 처리한 호선 정보와 같을 경우
        if(',' in row[-1]): #이용객 수를 str에서 int로 바꾸기 위해 ',' 문자 제거
            row[-1] = row[-1].replace(',','')
        if(',' in row[-2]):
            row[-2] = row[-2].replace(',','')
        row[-1] = int(row[-1])  #승하차 인원 수 계산하여 딕셔너리에 저장
        row[-2] = int(row[-2])
        tmp = row[-1] + row[-2]
        passengerCount[row[3]] = tmp
        prevLine = row[1]
        
    
```
